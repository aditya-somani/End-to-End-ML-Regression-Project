```python
import boto3
from pathlib import Path

# ----Config----
bucket = 'housing-regression-data-aditya-somani-ml-project' # name of the bucket
region = 'us-east-1'

# Set project root as parent of the notebooks folder
PROJECT_ROOT = Path.cwd().parent # get parent of current directory
local_data_dir = PROJECT_ROOT / "data" / "processed"
local_model_dir = PROJECT_ROOT / "models"

s3 = boto3.client("s3", region_name=region)
```

```python
help(s3.upload_file)
```

```python
# ----Helper Function----
def upload_file(
    local_path: Path,
    s3_key: str,
) -> None:
    if not local_path.exists():
        print(f"file not found at {local_path}")
        return
    print(f"Uploading {local_path} → s3://{bucket}/{s3_key}")
    s3.upload_file(str(local_path), bucket, s3_key)
```

```python
# ---- Upload required datasets ----
upload_file(local_data_dir / "feature_engineered_holdout.csv", "processed/feature_engineered_holdout.csv")
# upload_file(local_data_dir / "cleaning_holdout.csv", "processed_holdout.csv") - This line has been my error
upload_file(local_data_dir / "feature_engineered_train.csv", "processed/feature_engineered_train.csv")

# ---- Upload model ----
upload_file(local_model_dir / "xgb_best_model.pkl", "models/xgb_best_model.pkl")
```
